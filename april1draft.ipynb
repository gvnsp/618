# Data Wrangling

#### Hospital admissions
Data is number in thousands. We want rate. Need to multiply by 1000 and divide by population. Data is recorded every 5 years from 1950 to 1970. We will need to fill the missing values (pandas interpolation).

#### Cancer Death Rates
We have rates for each cancer type split by gender. Need to clean cancer types to include only cancers that both male and females have. Calculate combined rate (e.g. sum Stomach (male) & Stomach (female) = deaths/200,000 population)

#### Alcohol Expenditure
Divide by population to get expenditure per X population.

#### Air Pollution
We will use the per capita columns.

#### Join data
After the cleaning/wrangling we join dataframes by years

### Merging Dataframes

```python
# Merging dataframes

df1 = hospital_admissions_copy.merge(demographic_indicators_copy, on= 'Year')
df1 = df1.merge(cancer_death_copy, on = 'Year')
df1 = df1.merge(alcohol_expenditure_copy, on = 'Year')
df1 = df1.merge(air_pollution_copy, on = 'Year')
df1.head()
```

```python
df1.columns
```

```python
# # Converting the 'Total alcohol expenditure (USDA, 2018)' to rate by dividing by the 'Total Population, as of 1 July (thousands)' by year
# df1['Total alcohol expenditure rate'] = (df1['Total alcohol expenditure (USDA, 2018)'] / 100000)
# df1.head()
```

```python
# Stomach (combined) has values per 100,000 population in that year. Getting the total number of deaths by multiplying by the 'Total Population, as of 1 July (thousands)' by year
df1['Stomach (combined)'] = (df1['Stomach (combined)'] * df1['Total Population, as of 1 July (thousands)'])/100000
df1['Colon and Rectum (combined)'] = (df1['Colon and Rectum (combined)'] * df1['Female Population, as of 1 July (thousands)'])/100000
df1['Liver (combined)'] = (df1['Liver (combined)']  * df1['Female Population, as of 1 July (thousands)'])/100000
df1['Pancreas (combined)'] = (df1['Pancreas (combined)']  * df1['Female Population, as of 1 July (thousands)'])/10000
df1['Lung and Bronchus (combined)'] = (df1['Lung and Bronchus (combined)'] * df1['Female Population, as of 1 July (thousands)'])/10000
```

```python
df1.columns
```

- cancer combined columns have total numbers

```python
# Converting 'Total alcohol expenditure (USDA, 2018)' per 100k population
df1['Total alcohol expenditure (USDA, 2018)'] = (df1['Total alcohol expenditure (USDA, 2018)']/100000)
```
